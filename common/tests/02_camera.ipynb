Exercises the camera servo movements.
If this fails, it usually means ROS env not sourced for this kernel.

```python
from pathlib import Path
import sys

def add_repo_to_path():
    here = Path.cwd().resolve()
    for p in [here] + list(here.parents):
        if (p / 'lessons').is_dir() and (p / 'common').is_dir():
            repo_root = p
            common_lib = p / 'common' / 'lib'
            if str(repo_root) not in sys.path:
                sys.path.insert(0, str(repo_root))
            if str(common_lib) not in sys.path:
                sys.path.insert(0, str(common_lib))
            print('Repo root:', repo_root)
            print('Added common/lib:', common_lib)
            return
    raise FileNotFoundError('Could not find repo root (needs lessons/ and common/)')

add_repo_to_path()
```

```python
import time

import camera_lib

cam = camera_lib.get_camera()
print('Camera instance:', cam)

print('Center...')
cam.center_all(); time.sleep(0.5)

print('Nod...')
cam.nod(depth=220, speed_s=0.15); time.sleep(0.4)

print('Shake...')
cam.shake(width=220, speed_s=0.15); time.sleep(0.4)

print('Wiggle...')
cam.wiggle(cycles=1, amplitude=160, speed_s=0.12); time.sleep(0.4)

print('Center...')
cam.center_all()
print('Done')
```

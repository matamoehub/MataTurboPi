```python
import os, sys, importlib, time
mods = ("robot_moves","eyes_lib","camera_lib","tts_lib")
print("Python:", sys.version.split()[0])
for m in mods:
    try:
        __import__(m)
        print("✔ imported", m)
    except Exception as e:
        print("✖ import failed for", m, "->", e)
```

```python
import robot_moves as rm, importlib, time
importlib.reload(rm)
rm.set_base_speed(200); rm.set_rate(30)
print("Forward/back/turn/strafe/diagonal/drift…")
rm.forward(0.4); rm.stop(); time.sleep(0.2)
rm.backward(0.4); rm.stop(); time.sleep(0.2)
rm.turn_left(0.4); rm.stop(); time.sleep(0.2)
rm.turn_right(0.4); rm.stop(); time.sleep(0.2)
rm.left(0.35); rm.stop(); time.sleep(0.2)
rm.right(0.35); rm.stop(); time.sleep(0.2)
rm.diagonal_left(0.4); rm.diagonal_right(0.4); rm.stop(); time.sleep(0.2)
# wheel-level drift (uses set_motor_speeds stream)
if hasattr(rm, "drift_left_wheels"):
    rm.drift_left(0.6); rm.drift_right(0.6); rm.stop()
print("Movement demo complete.")
```

```python
import eyes_lib as elib, importlib, time
importlib.reload(elib)
eyes = elib.get_eyes() if hasattr(elib,"get_eyes") else getattr(elib,"eyes",None) or elib.Eyes()
eyes.set_both(255,0,0);  time.sleep(0.25)
eyes.set_both(0,255,0);  time.sleep(0.25)
eyes.set_both(0,0,255);  time.sleep(0.25)
eyes.blink(color=(253,208,0), period_s=0.18, duration_s=1.0)
eyes.wink("left",  color=(255,255,255), on_s=0.12, off_s=0.10, repeats=2)
eyes.wink("right", color=(255,255,255), on_s=0.12, off_s=0.10, repeats=2)
eyes.off()
print("Eyes demo complete.")
```

```python
import camera_lib as camlib, importlib, time
importlib.reload(camlib)
cam = camlib.get_cam() if hasattr(camlib,"get_cam") else getattr(camlib,"cam",None) or camlib.Camera()
cam.nod(); cam.shake(); cam.wiggle(cycles=1, amplitude=150); cam.center_all()
print("Camera demo complete.")
```

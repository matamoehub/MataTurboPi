## Build One Game Response
Choose one camera signal and one full game response.

Suggested flow:
- use blue lights for ready
- shake the camera head while the robot says `rock, paper, scissors`
- capture the hand sign
- say the player's move
- use green / yellow / red lights for result


````python
from lesson_header import *
import random
myRobot = bot(base_speed=300)

beats = ["Rock", "Paper", "Scissors", "Go"]
robot_move = random.choice(["rock", "paper", "scissors"])

myRobot.eyes.set_both(0, 0, 255)
myRobot.voice.say("Ready to play rock paper scissors", block=True)
myRobot.camera.center_all()
for word in beats:
    myRobot.voice.say(word, block=True)
    myRobot.camera.shake(width=180, speed_s=0.18)

hands = myRobot.vision.recognize_hands(show=True)
player_move = hands.get("game_moves", [None])[0]
print("robot move:", robot_move)
print("player move:", player_move)

def winner(player, robot):
    if player is None:
        return "retry"
    if player == robot:
        return "draw"
    wins = {("rock", "scissors"), ("paper", "rock"), ("scissors", "paper")}
    return "player" if (player, robot) in wins else "robot"

result = winner(player_move, robot_move)
if player_move is None:
    myRobot.eyes.set_both(255, 180, 0)
    myRobot.voice.say("I could not read your hand sign. Try again.", block=True)
elif result == "draw":
    myRobot.eyes.set_both(255, 180, 0)
    myRobot.voice.say(f"You played {player_move}. I played {robot_move}. It is a draw.", block=True)
elif result == "player":
    myRobot.eyes.set_both(255, 0, 0)
    myRobot.voice.say(f"You played {player_move}. I played {robot_move}. You win.", block=True)
else:
    myRobot.eyes.set_both(0, 255, 0)
    myRobot.voice.say(f"You played {player_move}. I played {robot_move}. I win.", block=True)

myRobot.camera.center_all()
````

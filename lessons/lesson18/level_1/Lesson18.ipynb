## Make The Robot Respond
Choose one camera signal and one robot response.

Suggested ideas:
- if a face is visible, say hello
- if the gesture is `thumbs_up`, beep
- if the pose is `hands_up`, change eye colour


````python
from lesson_header import *
myRobot = bot(base_speed=300)

faces = myRobot.vision.detect_faces(show=False)
hands = myRobot.vision.recognize_hands(show=False)
pose = myRobot.vision.detect_pose(show=False)

if faces["count"] > 0:
    myRobot.voice.say("I can see a face", block=True)

if hands["hands"] and hands["hands"][0]["gesture"] == "thumbs_up":
    myRobot.buzzer.beep(note="C5", duration_s=0.15)

if pose["label"] == "hands_up":
    myRobot.eyes.set_both(0, 255, 0)
````

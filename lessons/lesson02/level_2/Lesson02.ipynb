## Step 2 — Build your sequence function

Checklist:
- async blink + movement
- at least one wink
- camera nod + shake + look up/left/down/right
- fidget for at least 3 seconds at 1/4 speed or slower
- horn normal once


```python
def my_sequence(anim):
    anim.set_eye_color((80, 220, 255))
    anim.start_blinking(every_s=2.8, blank_s=0.5)
    anim.start_fidget(step_s=0.05, speed_scale=0.25, seed=22)

    try:
        anim.nod(depth=220)
        anim.play_phrase('line_1', block=True)

        anim.wink('left', blank_s=0.5)
        anim.look_up(amplitude=220)
        anim.look_left(amplitude=220)
        anim.look_down(amplitude=220)
        anim.look_right(amplitude=220)

        anim.play_phrase('line_2', block=True)
        anim.shake(width=250)
        anim.play_phrase('line_3', block=True)
    finally:
        anim.stop_fidget()
        anim.stop_blinking()

    anim.robot.drift_left(seconds=1.0, turn_blend=0.95)
    anim.robot.drift_right(seconds=1.0, turn_blend=0.95)
    anim.horn_normal()
```

```python
my_sequence(anim)
```

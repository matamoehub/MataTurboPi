```python
from lesson_header import *
myRobot = bot(base_speed=300)
import time

try:
    print("Eyes: red, green, blue, off")
    myRobot.eyes.set_both(255, 0, 0); time.sleep(0.4)
    myRobot.eyes.set_both(0, 255, 0); time.sleep(0.4)
    myRobot.eyes.set_both(0, 0, 255); time.sleep(0.4)
    myRobot.eyes.off(); time.sleep(0.2)
    print("Eyes: blink white")
    myRobot.eyes.blink_once(blank_s=0.3)
    print("Eyes demo done")
except Exception as e:
    print("Eyes demo failed:", e)
```

```python
from lesson_header import *
myRobot = bot(base_speed=300)
import time

try:
    print("Center...")
    myRobot.camera.center_all()
    print("Nod...")
    myRobot.camera.nod(depth=250)
    print("Shake...")
    myRobot.camera.shake(width=250)
    print("Wiggle...")
    myRobot.camera.wiggle(cycles=2, amplitude=200)
    print("Tiny wiggle...")
    myRobot.camera.tiny_wiggle(seconds=2.0, amplitude=90, speed_s=0.12)
    print("Camera head demo done")
except Exception as e:
    print("Camera head demo failed:", e)
```

```python
from lesson_header import *
myRobot = bot(base_speed=300)
import time

try:
    print("Forward / Backward")
    myRobot.move.forward(0.6); time.sleep(0.25)
    myRobot.move.backward(0.6); time.sleep(0.25)
    myRobot.stop(); time.sleep(0.1)

    print("Turn left / right")
    myRobot.move.turn_left(0.5); time.sleep(0.25)
    myRobot.move.turn_right(0.5); time.sleep(0.25)
    myRobot.stop(); time.sleep(0.1)

    print("Strafe left / right")
    myRobot.move.left(0.6); time.sleep(0.25)
    myRobot.move.right(0.6); time.sleep(0.25)
    myRobot.stop(); time.sleep(0.1)

    print("Movement demo done")
except Exception as e:
    myRobot.stop()
    print("Movement failed:", e)
```

```python
from lesson_header import *
myRobot = bot(base_speed=300)
import time

try:
    print("Drift left / right")
    myRobot.move.drift_left(seconds=0.9, turn_blend=0.55)
    time.sleep(0.2)
    myRobot.move.drift_right(seconds=0.9, turn_blend=0.55)
    time.sleep(0.2)
    myRobot.stop()
    print("Drift demo done")
except Exception as e:
    myRobot.stop()
    print("Drift failed:", e)
```

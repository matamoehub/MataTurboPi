# Level 1 - Single Cup Navigation

Write a series of instructions that will allow your robot to drive a circle around a singular placed cup. While this can be achieved simply with basic movements, your task is to navigate around the cup with some style. Be creative in how you choose to make your robot move!

### Available commands:

#### Moving
Moves the robot in the selected direction regardless of its current rotation
The number in the parentheses is the movement duration in seconds. The value 0.5 used here is just an example, you will want to change this value when writing your commands
```Python
myRobot.move.forward(0.5)
myRobot.move.backward(0.5)
myRobot.move.left(0.5)
myRobot.move.right(0.5)
```
#### Turning
Rotates the robot left or right for a given number of seconds. If you want your robot to rotate specific angles you will need to play around with the duration to see how much it turns!
```Python
myRobot.move.turn_left(0.5)
myRobot.move.turn_right(0.5)
```
#### Drifting
Drift commands have an optional additional parameter for how curved the drift is (0.0 = mostly sideways, 1.0 = strong turning arc)
```Python
myRobot.move.drift_left(0.5)
myRobot.move.drift_right(0.5)
myRobot.move.drift_right(0.5, turn_blend=0.55) 
```
#### Extra
```Python
myRobot.stop() # stops the robot
myRobot.horn() # (meep meep)
time.sleep(0.15) # waits the amount of time in seconds
```

# Matamoe Robot League

Robot league is comprised of four different levels, each of increasing difficulty. In Jupyter Notebook, code is run in 'code blocks'. Your task is to write code inside code blocks that when run, will complete the current level. Each level will be introduced sequentially, and an empty code block created for you to write your code in.

Robot league is comprised of eight classes. After each Level, a conjoint session will be held where you will compete in a challenge using the skills your learnt in the previous level!

### Robot commands

Each level will require you to write a series of commands that will control the robot.

To write a robot command, we first need to get a reference to our robot.
This can be done like so:

```Python
myRobot = bot(base_speed = 300)
```
300 is a good base value for speed, but feel free to play around with this. myRobot is just the name of the variable, you can set this to anything you want if you want to name your robot!

With a reference to our robot, commands can be written as:
```Python
myRobot.myCommand()
```
All of the necessary commands for a given level will be provided above the level work space, so you can always copy and paste the commands.

Finally, importing a header file is necessary within each code block. You will need to paste the following line at the top of each of your level workspaces for the code to work properly:
```Python
from lesson_header import *
```

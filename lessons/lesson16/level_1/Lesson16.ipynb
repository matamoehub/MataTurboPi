## For Loop Example - Iterate Through Eye Colours

This demonstrates iteration through a list.
Each obstacle number changes both the colour and the number of flashes.

```python
from lesson_header import *
myRobot = bot(base_speed=300)

eye_colours = [
    (255, 0, 0),
    (0, 255, 0),
    (0, 0, 255),
    (255, 255, 0),
]

obstacle_number = 2
colour = eye_colours[obstacle_number - 1]

for i in range(obstacle_number):
    myRobot.eyes.set_both(*colour)
    time.sleep(0.25)
    myRobot.eyes.off()
    time.sleep(0.25)
```

## Function Example 2

This function combines a variable, a list, and a loop.

```python
from lesson_header import *
myRobot = bot(base_speed=300)

def flash_progress(obstacle_number, colour_list):
    colour = colour_list[(obstacle_number - 1) % len(colour_list)]
    for i in range(obstacle_number):
        myRobot.eyes.set_both(*colour)
        time.sleep(0.2)
        myRobot.eyes.off()
        time.sleep(0.2)

flash_progress(3, [(255, 0, 0), (0, 255, 0), (0, 0, 255)])
```

## If / Else Example 1

Use sonar distance to choose an action.

```python
from lesson_header import *
myRobot = bot(base_speed=300)

distance_cm = myRobot.sonar.distance_cm(filtered=True)
print('distance_cm =', distance_cm)

if distance_cm is None:
    myRobot.voice.say('No sonar reading yet.')
elif distance_cm < 20:
    myRobot.voice.say('Obstacle close.')
    myRobot.eyes.set_both(255, 0, 0)
else:
    myRobot.voice.say('Path is clear.')
    myRobot.eyes.set_both(0, 255, 0)

time.sleep(0.5)
myRobot.eyes.off()
```

## If / Else Example 2

Choose a movement based on how far away the obstacle is.

```python
from lesson_header import *
myRobot = bot(base_speed=300)

distance_cm = myRobot.sonar.distance_cm(filtered=True)

if distance_cm is None:
    print('No reading yet')
elif distance_cm < 20:
    myRobot.move.turn_left(0.4)
elif distance_cm < 35:
    myRobot.move.left(0.4)
else:
    myRobot.move.forward(0.4)

myRobot.stop()
```

## While Loop Example

A `while` loop can repeat a pattern multiple times.
For example, this can help the robot go around a cup more than once.

```python
from lesson_header import *
myRobot = bot(base_speed=300)

laps = 0
max_laps = 2

while laps < max_laps:
    myRobot.move.forward(0.4)
    myRobot.move.turn_left(0.4)
    laps = laps + 1

myRobot.stop()
print('laps completed:', laps)
```

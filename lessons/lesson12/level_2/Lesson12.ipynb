## Step 2 — Build your sequence function

Checklist:
- async blink + movement
- at least one wink
- camera nod + shake + look up/left/down/right
- fidget for at least 3 seconds at 1/4 speed or slower
- horn normal once

```python
from lesson_header import *
myRobot = bot(base_speed=300)

def my_sequence(robot):
    robot.anim.set_eye_color((80, 220, 255))
    robot.anim.start_blinking(every_s=2.8, blank_s=0.5)
    robot.anim.start_fidget(step_s=0.05, speed_scale=0.25, seed=22)

    try:
        robot.anim.nod(depth=220)
        robot.voice.play_phrase('line_1', block=True)

        robot.anim.wink('left', blank_s=0.5)
        robot.anim.look_up(amplitude=220)
        robot.anim.look_left(amplitude=220)
        robot.anim.look_down(amplitude=220)
        robot.anim.look_right(amplitude=220)

        robot.voice.play_phrase('line_2', block=True)
        robot.anim.shake(width=220)
        robot.anim.horn_normal()

        robot.voice.play_phrase('line_3', block=True)
        robot.move.drift_left(1.0)
        robot.move.drift_right(1.0)
    finally:
        robot.anim.stop()
```

```python
from lesson_header import *
myRobot = bot(base_speed=300)
my_sequence(myRobot)
```

## Async vs Sync Quick Guide

- `myRobot.move.forward(...)` waits until complete (**sync**)
- `myRobot.anim.move_async(...)` starts and returns immediately (**async**)
- `myRobot.eyes.start_blink(...)` runs in the background (**async**)
- `myRobot.anim.start_fidget(...)` runs in the background (**async**)
- `myRobot.voice.play_phrase(..., block=True)` waits (**sync**)
- `myRobot.voice.play_phrase(..., block=False)` does not wait (**async**)

This matters because animation often needs more than one thing happening at once.

## Example 1 — Async move + blink
Make the robot move while the eyes keep blinking.

This is one of the first times students see why asynchronous actions matter.

```python
from lesson_header import *
myRobot = bot(base_speed=300)
myRobot.eyes.start_blink(every_s=2.5, blank_s=0.5)
myRobot.anim.move_async('forward', seconds=1.0, speed=180)
myRobot.move.backward(0.8)
myRobot.eyes.stop_blink()
myRobot.stop()
```

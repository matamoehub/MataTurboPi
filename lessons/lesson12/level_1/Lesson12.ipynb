## Async vs Sync Quick Guide
- `myRobot.move.forward(...)` waits until the move ends
- `myRobot.anim.move_async(...)` starts a move and returns immediately
- `myRobot.eyes.blink(...)` starts background blinking
- `myRobot.anim.start_fidget(...)` starts background fidgeting
- `myRobot.voice.play_phrase(..., block=True)` waits
- `myRobot.voice.play_phrase(..., block=False)` does not wait


## Example 1 — Async move + blink
Make the robot move while the eyes keep blinking.


```python
from lesson_header import *
myRobot = bot(base_speed=300)
myRobot.eyes.color(0, 220, 255)
myRobot.eyes.blink(every_s=2.5, blank_s=0.5)
myRobot.anim.move_async("forward", seconds=1.0, speed=180)
myRobot.move.backward(0.8, speed=160)
myRobot.eyes.stop_blink()
```

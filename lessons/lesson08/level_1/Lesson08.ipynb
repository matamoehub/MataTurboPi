## Build One Game Response
Start with one small test before building the whole game.

Suggested ideas:
- if the player shows `fist`, say `Rock`
- if the player shows `open_palm`, say `Paper`
- if the player shows `peace`, say `Scissors`
- if the player is not framed well, ask them to stand in front of the camera


````python
from lesson_header import *

faces = vision.detect_faces(show=False)
hands = vision.recognize_hands(show=False)

if faces["count"] == 0:
    tts.say("Stand in front of the camera to play", block=True)
elif hands["hands"]:
    gesture = hands["hands"][0]["gesture"]
    if gesture == "fist":
        tts.say("Rock", block=True)
    elif gesture == "open_palm":
        tts.say("Paper", block=True)
    elif gesture == "peace":
        tts.say("Scissors", block=True)
    else:
        tts.say(f"I saw {gesture}", block=True)
else:
    tts.say("Show me your hand sign", block=True)
````

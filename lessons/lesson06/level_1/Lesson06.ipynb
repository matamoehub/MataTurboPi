## If / Else Example 1

## Project Skeleton

Use this as the base for your obstacle-course performance.
You should expand it and make it creative.

```python
from lesson_header import *

number_of_obstacles = 3
obstacle_count = 0
eye_colours = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
warning_distance_cm = 25


def flash_progress(obstacle_number):
    colour = eye_colours[(obstacle_number - 1) % len(eye_colours)]
    for i in range(obstacle_number):
        eyes.set_both(*colour)
        time.sleep(0.2)
        eyes.off()
        time.sleep(0.2)


def react_to_distance(distance_cm):
    if distance_cm is None:
        tts.say('Waiting for sonar')
    elif distance_cm < warning_distance_cm:
        tts.say('Too close')
        bz.beep(note='C4', duration_s=0.15)
        moves.turn_left(0.4)
    else:
        moves.forward(0.4)


tts.say('Starting obstacle course')
cam.center_all()

while obstacle_count < number_of_obstacles:
    distance_cm = sonar.get_distance_cm(filtered=True)
    react_to_distance(distance_cm)
    obstacle_count = obstacle_count + 1
    flash_progress(obstacle_count)

stop_project_robot()
tts.say('Course complete')
```

# Teacher Solution — Lesson 1 (Single Cup)

This is a **reference path** around one cup.
Adjust timings depending on floor grip + battery.

Uses **base** library: `common/lib/robot_moves.py`

```python
from pathlib import Path
import sys, importlib.util, time

# Find repo root by walking up until we see common/ and lessons/
here = Path.cwd().resolve()
ROOT = None
for p in [here] + list(here.parents):
    if (p/'common').is_dir() and (p/'lessons').is_dir():
        ROOT = p
        break
if ROOT is None:
    raise FileNotFoundError('Repo root not found (need common/ and lessons/)')

COMMON_LIB = ROOT/'common'/'lib'
if str(COMMON_LIB) not in sys.path:
    sys.path.insert(0, str(COMMON_LIB))

def import_base_robot_moves():
    path = COMMON_LIB / 'robot_moves.py'
    spec = importlib.util.spec_from_file_location('base_robot_moves', str(path))
    mod = importlib.util.module_from_spec(spec)
    assert spec and spec.loader
    spec.loader.exec_module(mod)
    return mod

rm = import_base_robot_moves()
rm.set_base_speed(220)
rm.set_rate(30)
print('Using base robot_moves:', rm.__file__)
```

## Path plan

Simple rectangle around cup:
- forward
- strafe right
- forward
- strafe left
- forward

Start + finish with horn.

```python
rm.horn(block=True)
time.sleep(0.2)

# Approach cup
rm.forward(0.7); rm.stop(); time.sleep(0.2)

# Go around right side of cup
rm.right(0.55); rm.stop(); time.sleep(0.2)

# Past the cup
rm.forward(0.8); rm.stop(); time.sleep(0.2)

# Come back to original line
rm.left(0.55); rm.stop(); time.sleep(0.2)

# Finish stretch
rm.forward(0.6); rm.stop(); time.sleep(0.2)

rm.horn(block=True)
print('Done ✅')
```

## Emergency stop

```python
rm.emergency_stop(); print('Stopped')
```
